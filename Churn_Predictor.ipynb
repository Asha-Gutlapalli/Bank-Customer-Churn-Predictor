## Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle

from keras.models import Sequential
from keras.layers import Dense, Dropout
```

## Data Loading

```python
dataset = pd.read_csv('data/Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
Y = dataset.iloc[:, 13].values

dataset
```

## Encoding

```python
Encoder_X1 = LabelEncoder()
X[:, 1] = Encoder_X1.fit_transform(X[:, 1])

Encoder_X2 = LabelEncoder()
X[:, 2] = Encoder_X2.fit_transform(X[:, 2])

X.shape
```

## Train-Test Split

```python
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

## Scaling

```python
sc = StandardScaler()

Train_X = sc.fit_transform(Train_X)
Test_X = sc.transform(Test_X)
```

## Model

```python
classifier = Sequential()
classifier.add(Dense(units = 32, activation = 'relu', input_dim = 10))
classifier.add(Dense(units = 16, activation = 'relu'))
classifier.add(Dense(units = 8, activation = 'relu'))
classifier.add(Dropout(0.4))
classifier.add(Dense(units = 1, activation = 'sigmoid'))

classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

## Train

```python
classifier.fit(Train_X, Train_Y, batch_size = 10, epochs = 200)
```

## Save Model

```python
classifier.save_weights("models/model")
```

## Save Scaler

```python
with open('scalers/scaler.pickle', 'wb') as handle:
    pickle.dump(sc, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

## Save Encoders

```python
with open('encoders/encoder1.pickle', 'wb') as handle:
    pickle.dump(Encoder_X1, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
